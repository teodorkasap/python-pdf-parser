```python
# Find the row where the table starts
for row in range(df.shape[0]):
    for col in range(df.shape[1]):
        if df.iat[row, col] == 'Payment:':
            row_start = row
            break

# Find the row where the table ends
for row in range(df.shape[0]):
    for col in range(df.shape[1]):
        if df.iat[row, col] == 'Summary:':
            row_end = row
            break

df = df.loc[row_start:row_end]
df.columns = range(df.shape[1])
df.drop(df.tail(1).index, inplace=True)
df = df.iloc[3:, [0, 1, 2, 3, 4, 5, 6]]
df = df.fillna(0)
df = df[(df[[0]] != 0).all(axis=1)]
df
```
